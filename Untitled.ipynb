```python
temizlenmisVeri = temizlenmisVeri[temizlenmisVeri.year != 1970]        # 1970 model araçların verisini temizleme
temizlenmisVeri.groupby("year").mean(numeric_only=True)["price"]
```

```python
temizlenmisVeri = temizlenmisVeri.drop(["model", "transmission", "fuelType"], axis=1)    # numeric bir değer olmadığı için hata almamak adına veriyi silelim.
temizlenmisVeri
```

```python
y = temizlenmisVeri["price"].values
x = temizlenmisVeri.drop("price", axis=1).values
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
model = Sequential()
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))
model.add(Dense(14,activation="relu"))



model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), batch_size=250, epochs=500)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
tahmin = model.predict(x_test)
```

```python
tahmin
```

```python
mean_absolute_error(y_test, tahmin)
```
